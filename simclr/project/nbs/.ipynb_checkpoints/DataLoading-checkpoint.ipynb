```python
path=untar_data(URLs.IMAGENETTE_320)
```

The very first thing we have to do is make sure that we have two copies of each image. This copy is done on the GPU in order to avoid data transfer cost between cpu and gpu. 

```python
class CopyImage(Transform):
    def __init__(self,copies=2):
        self.copies=copies
    def __call__(self,b,**kwargs):
        return (b[0].repeat_interleave(self.copies,0),b[1].repeat_interleave(self.copies,0))
```

```python
imagenet = DataBlock(blocks = (ImageBlock(),CategoryBlock),
                  get_items = get_image_files,
                  splitter = GrandparentSplitter(),
                  item_tfms=CropPad(340),
                  batch_tfms=CopyImage(),
                  get_y = parent_label)
```

```python
dls = imagenet.dataloaders(path)
```
